```python
import os
import shutil
import random
random.seed(18)
carpeta_imagenes = '/content/seeds'

carpeta_destino = '/content/seeds_20'
os.makedirs(carpeta_destino, exist_ok=True)

imagenes = os.listdir(carpeta_imagenes)

imagenes = [img for img in imagenes if img.endswith(('.JPEG'))]

rand_70 = random.sample(imagenes, 40)

for imagen in rand_70:
    shutil.copy(os.path.join(carpeta_imagenes, imagen), carpeta_destino)
```

```python
root_folder = '/content/seeds_20'

image_data = []
for root, dirs, files in os.walk(root_folder):
    for file in files:
        if file.endswith(".JPEG"):
            full_path = os.path.join(root, file)
            image_data.append(full_path)
print(image_data)

for model_name in ["vgg16", "vgg19", "resnet50"]:
  for img_path in image_data:
      command = f"python2.7 /content/TFG/Grad-CAM/grad-cam_imagenet_vgg-resnet.py \"{img_path}\" \"{model_name}\""
      !{command}
```

```python
import os

root_folder = '/content/MultiSignDB/'

image_extensions = ('.png', '.jpg', '.jpeg', '.bmp', '.gif')

image_data = []

for root, dirs, files in os.walk(root_folder):
    for file in files:
        if file.lower().endswith(image_extensions):
            full_path = os.path.join(root, file)
            relative_path = os.path.relpath(full_path, root_folder)
            parts = relative_path.split(os.sep)
            if len(parts) >= 2:
                second_folder = parts[0]
            else:
                second_folder = "unknown"
            image_data.append((full_path, second_folder))
for model_name in ["model1", "model2", "model3"]:
  for img_path, folder in image_data:
      command = f"python2.7 /content/TFG/Grad-CAM/grad-cam_multisign.py \"{img_path}\" \"{model_name}\" \"{folder}\""
      !{command}
```
