```python
import pickle
import numpy as np

def convert_numpy_dict(d):
    """Recursively converts all NumPy arrays to standard Python lists"""
    if isinstance(d, dict):
        return {k: convert_numpy_dict(v) for k, v in d.items()}
    elif isinstance(d, list):
        return [convert_numpy_dict(v) for v in d]
    elif isinstance(d, np.ndarray):
        return d.tolist()
    else:
        return d

# Load original pickle
with open('/content/mini-imagenet/mini-imagenet-cache-train.pkl', 'rb') as f:
    obj = pickle.load(f)

# Convert all NumPy arrays to standard types
clean_obj = convert_numpy_dict(obj)

# Re-pickle using protocol=2
with open('/content/mini-imagenet/mini-imagenet-cache-train-py2.pkl', 'wb') as f:
    pickle.dump(clean_obj, f, protocol=2)
```
